### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 2. Ventas por categoría

```python
categorias_tienda =[]
conteo_categorias_tienda ={}
categorias_tienda2 =[]
conteo_categorias_tienda2 ={}
categorias_tienda3 =[]
conteo_categorias_tienda3 ={}
categorias_tienda4 =[]
conteo_categorias_tienda4 ={}

# Tienda 1
for categoria in tienda['Categoría del Producto']:
    if categoria not in categorias_tienda:
        categorias_tienda.append(categoria)
        conteo_categorias_tienda[categoria] = 1
    else:
        conteo_categorias_tienda[categoria] += 1
print("Tienda 1:")

for categoria, cantidad in sorted(conteo_categorias_tienda.items(), key=lambda x: x[1], reverse=True):
    print(f"{categoria}: {cantidad}")
print("\n")

# Tienda 2
for categoria in tienda2['Categoría del Producto']:
    if categoria not in categorias_tienda2:
        categorias_tienda2.append(categoria)
        conteo_categorias_tienda2[categoria] = 1
    else:
        conteo_categorias_tienda2[categoria] += 1
print("Tienda 2:")
for categoria, cantidad in sorted(conteo_categorias_tienda2.items(), key=lambda x: x[1], reverse=True):
    print(f"{categoria}: {cantidad}")
print("\n")

# Tienda 3
for categoria in tienda3['Categoría del Producto']:
    if categoria not in categorias_tienda3:
        categorias_tienda3.append(categoria)
        conteo_categorias_tienda3[categoria] = 1
    else:
        conteo_categorias_tienda3[categoria] += 1
print("Tienda 3:")
for categoria, cantidad in sorted(conteo_categorias_tienda3.items(), key=lambda x: x[1], reverse=True):
    print(f"{categoria}: {cantidad}")
print("\n")

# Tienda 4
for categoria in tienda4['Categoría del Producto']:
    if categoria not in categorias_tienda4:
        categorias_tienda4.append(categoria)
        conteo_categorias_tienda4[categoria] = 1
    else:
        conteo_categorias_tienda4[categoria] += 1
print("Tienda 4:")
for categoria, cantidad in sorted(conteo_categorias_tienda4.items(), key=lambda x: x[1], reverse=True):
    print(f"{categoria}: {cantidad}")
```

```python
conteo_categorias_total = {}
todas_las_tiendas = [tienda, tienda2, tienda3, tienda4]

for t in todas_las_tiendas:
    for categoria in t['Categoría del Producto']:
        if categoria in conteo_categorias_total:
            conteo_categorias_total[categoria] += 1
        else:
            conteo_categorias_total[categoria] = 1

print("Categorías más populares:")
for categoria, cantidad in sorted(conteo_categorias_total.items(), key=lambda x: x[1], reverse=True):
    print(f"{categoria}: {cantidad}")
```

```python
categorias_ordenadas = sorted(conteo_categorias_total.items(), key=lambda x: x[1], reverse=True)
categorias = [x[0] for x in categorias_ordenadas]
conteos = [x[1] for x in categorias_ordenadas]


plt.figure(figsize=(10, 6))
plt.barh(categorias, conteos, color= 'm')
plt.xlabel("Cantidad de productos")
plt.title("Categorías más populares")
plt.gca().invert_yaxis()  # Poner la categoría más popular arriba
plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda


```python
tiendas =['Tienda','Tienda2','Tienda3','Tienda4']

sum_calificacion_tienda = tienda['Calificación'].sum()
cuenta_calificaciones_tienda = tienda['Calificación'].count()

sum_calificacion_tienda2 = tienda2['Calificación'].sum()
cuenta_calificaciones_tienda2 = tienda2['Calificación'].count()

sum_calificacion_tienda3 = tienda3['Calificación'].sum()
cuenta_calificaciones_tienda3 = tienda3['Calificación'].count()

sum_calificacion_tienda4 = tienda4['Calificación'].sum()
cuenta_calificaciones_tienda4 = tienda4['Calificación'].count()

def promedio_calificacion( sum , cuenta):
  promedio = sum / cuenta
  return promedio

promedio_tienda = promedio_calificacion(sum_calificacion_tienda, cuenta_calificaciones_tienda)

promedio_tienda2 = promedio_calificacion(sum_calificacion_tienda2, cuenta_calificaciones_tienda2)

promedio_tienda3 = promedio_calificacion(sum_calificacion_tienda3, cuenta_calificaciones_tienda3)

promedio_tienda4 = promedio_calificacion(sum_calificacion_tienda4, cuenta_calificaciones_tienda4)

print(tiendas)
print(promedio_tienda)
print(promedio_tienda2)
print(promedio_tienda3)
print(promedio_tienda4)
```

```python
import matplotlib.pyplot as plt

tiendas =['Tienda','Tienda2','Tienda3','Tienda4']
promedios = [promedio_tienda, promedio_tienda2, promedio_tienda3, promedio_tienda4]

plt.figure(figsize=(8, 5))
plt.plot(tiendas, promedios, marker='o', linestyle='-', color='m')
plt.title("Promedio de Calificación por Tienda")
plt.xlabel("Tienda")
plt.ylabel("Promedio de Calificación")
plt.ylim(0, 5)
plt.grid(True)
plt.tight_layout()
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

def top_productos_grafica(tienda, nombre_tienda):
    productos_tienda = []
    conteo_productos_tienda = {}

    for producto in tienda['Producto']:
        if producto not in productos_tienda:
            productos_tienda.append(producto)
            conteo_productos_tienda[producto] = 1
        else:
            conteo_productos_tienda[producto] += 1

    productos_ordenados = sorted(conteo_productos_tienda.items(), key=lambda x: x[1], reverse=True)

    print(f"\nTop 10 productos más vendidos en {nombre_tienda}:")
    for producto, cantidad in productos_ordenados[:10]:
        print(f"{producto}: {cantidad}")

    print(f"\nTop 10 productos menos vendidos en {nombre_tienda}:")
    for producto, cantidad in productos_ordenados[-10:]:
        print(f"{producto}: {cantidad}")

    top_10_productos = productos_ordenados[:10]
    productos = [item[0] for item in top_10_productos]
    cantidades = [item[1] for item in top_10_productos]

    plt.figure(figsize=(8, 8))
    plt.pie(cantidades, labels=productos, autopct='%1.1f%%', startangle=140)
    plt.title(f'Top 10 Productos Más Vendidos - {nombre_tienda}')
    plt.axis('equal')
    plt.tight_layout()
    plt.show()

top_productos_grafica(tienda, "Tienda 1")
top_productos_grafica(tienda2, "Tienda 2")
top_productos_grafica(tienda3, "Tienda 3")
top_productos_grafica(tienda4, "Tienda 4")
```
